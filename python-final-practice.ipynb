```python
x = np.array([0, 1, 2, 3, 4])  # Values
freq = np.array([32, 178, 290, 236, 64])  # Observed frequencies
n = 4  # Total number of trials
p = 0.5  # Probability of success

# Step 1: Compute expected frequencies using binomial PMF
pmf = binom.pmf(x, n, p)  # Binomial probabilities
expected = freq.sum() * pmf  # Scale probabilities to total observations

print(pmf)

x_cal = chisquare(f_obs=freq, f_exp=expected).statistic

x_tab = scipy.stats.chi2.ppf(0.95, df=len(x) - 1)  # Degrees of freedom = categories - 1

print(f"x_cal: {x_cal}\nx_tab: {x_tab}")
if x_cal < x_tab:
    print("Accept: The data follows a binomial distribution.")
else:
    print("Reject: The data does not follow a binomial distribution.")
```
